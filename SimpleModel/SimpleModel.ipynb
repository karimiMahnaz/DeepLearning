```python
import numpy as np
import tensorflow as tf

from tensorflow import keras
from keras import layers
```

Data Definition

```python
# y = 3 * x-5

x= tf.constant([-1,-2,0,1,2,5,7], dtype=float)
y=3*x-5
```

```python
x.numpy()
```

```python
y.numpy()
```

Model Definition

```python
model = keras.Sequential([

    keras.layers.Dense(units = 1, activation=None, input_shape=[1])
])
```

```python
# (y-y_pred)^2

model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
model.summary()
```

```python
model.fit(x, y, batch_size= 1, epochs=500)
```

```python
x_test = tf.constant([-4, 11 , 20], dtype=float)
y_test = x_test * 3 - 5
```

```python
y_test.numpy()
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x_test.numpy(), y_test.numpy(), color='red', linewidths=6)
plt.scatter(x_test.numpy(), y_pred, color='blue', linewidths=3)
plt.show()
```
